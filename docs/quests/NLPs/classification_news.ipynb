```python
import pandas as pd
df_IT_News = pd.read_excel('IT_News.xlsx', sheet_name=None, engine='openpyxl')
```

```python
df_sheets = []
for sheet_name, df in df_IT_News.items() :
  df_sheets.append(df)
  pass
df_sheets[0]
```

```python
for x in df_sheets :
  if x.columns[0] != '이름' :
    x.rename(columns = {f'{x.columns[0]}':'이름'}, inplace=True)
    pass
df_sheets
```

```python
news_name = []
for x in df_sheets :
  news_name = x['뉴스명'].dropna().values.tolist()+news_name
news_name
```

```python
# 숫자 삭제, 영문-소문자 통일
news_name = pd.Series(news_name)
news_name = news_name.str.replace('[0-9]','',regex=True).str.lower()
news_name
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
# 조사, 어미, 구두점 같은 것을 제거
def okt_clean(text):
    clean_text = []
    for word in okt.pos(text, stem=True):
        if word[1] not in ['Josa', 'Eomi', 'Punctuation','URL']:
            clean_text.append(word[0])

    return " ".join(clean_text)
```

```python
okt.pos('https://www.dailian.co.kr/news/view//?sc=Naver', stem=True)
```

```python
news_name = news_name.apply(okt_clean)
news_name
```

```python
file_path = 'tfidfVectorizer.pkl'

import pickle
with open(file_path, 'rb') as pickle_file:
  tfidfVectorizer = pickle.load(pickle_file)
```

```python
# Tokenizing
sentence_tokens = tfidfVectorizer.transform(news_name) # 학습 형식 그대로 이용
sentence_tokens
```

```python
file_path = 'NLP_classfication_ynanewstitles.pkl'

import pickle
with open(file_path, 'rb') as pickle_file:
  loaded_model_cate = pickle.load(pickle_file)
```

```python
predict_cate = loaded_model_cate.predict(news_name)
```

```python
df_topic = pd.read_csv('topic_dict.csv')
df_topic
```

```python
df_topic.query(f"topic_idx==2").values[0][0]
```

```python
predict_topic = []
for x in predict_cate :
  topic = df_topic.query(f"topic_idx=={x}").values[0][0]
  predict_topic.append(topic)
predict_topic
```
